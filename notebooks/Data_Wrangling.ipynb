```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/workspaces/EDA-on-Netflix-Dataset/datasets/raw/netflix_movies_detailed_up_to_2025.csv")
```

```python
df = df.drop(['duration'], axis=1)
```

```python
df.columns
```

```python
# Dropping all the rows which have the NaN values (Columns with names or alphabetical stuffs).
df = df.dropna(subset=['director', 'cast', 'country', 'description', 'genres'])
```

```python
# Finally check full data if any of the Null valued rows left.
df.isnull().sum()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
# Labelled encoded the 'language' column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['language'] = le.fit_transform(df['language'])
```

```python
# Final checking of is there anything that can be changed to int or float
df.info()
```

```python
# Saving the cleaned dataframe

df.to_csv("/workspaces/EDA-on-Netflix-Dataset/datasets/processed/netflix_movies_processed.csv")
```
